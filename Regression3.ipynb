```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
import scipy
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('bodym.csv')
df = df[df.gender == 'male']
```

```python
L_columns = df.columns.tolist()
L_columns.remove('subject_id')
L_columns.remove('gender')
L_columns.remove('photo_names')
L_columns

for col in L_columns:
    L = L_columns.copy()
    L.remove(col)
    X = df[L]
    Y = df[col]
    X_train = X.iloc[:2*len(X)//3]
    Y_train = Y.iloc[:2*len(Y)//3]
    X_test = X.iloc[len(X)//2:]
    Y_test = Y.iloc[len(Y)//2:]

    print("")

    print("moyenne ",col," : ",Y.mean())

    reg = LinearRegression()
    reg.fit(X_train, Y_train)

    print("root mean squared error")
    rmse_reg_train = mean_squared_error(Y_train, reg.predict(X_train), squared=False)
    print("\r train: {0:0.2f}".format(rmse_reg_train))

    rmse_reg_test = mean_squared_error(Y_test, reg.predict(X_test), squared=False)
    print("\r test: {0:0.2f}".format(rmse_reg_test))
        
```

```python
#N'affiche que les trois premiers coefficients les plus importants
#Version standaridisée

#Nous créons également un dico contenant le nombre de fois que les variables apparaissent dans les trois premiers coefficients
dico = {}



L_columns = df.columns.tolist()
L_columns.remove('subject_id')
L_columns.remove('gender')
L_columns.remove('photo_names')
L_columns

#Initailisation du dico
for col in L_columns:
    dico[col] = 0

for col in L_columns:
    L = L_columns.copy()
    L.remove(col)
    X = np.array(df[L])
    n = X.shape[0]
    Y = np.array(df[col])
    X_train = X[:2*n//3, :]
    Y_train = Y[:2*n//3]
    # Standardisation de X_train
    stdX = StandardScaler()
    stdX.fit(X_train)
    X_train = stdX.transform(X_train)
    X_test = X[2*n//3:, :]
    X_test = stdX.transform(X_test)
    Y_test = Y[2*n//3:]

    
    reg = LinearRegression()
    reg.fit(X_train, Y_train)

    Coeff = reg.coef_.copy()
    print("\nCoefficients de ",col,":")
    i1 = np.argmax(np.abs(Coeff))
    Coeff[i1] = 0
    i2 = np.argmax(np.abs(Coeff))
    Coeff[i2] = 0
    i3 = np.argmax(np.abs(Coeff))
    print(L[i1],": ",reg.coef_[i1])
    dico[L[i1]] += 1
    print(L[i2],": ",reg.coef_[i2])
    dico[L[i2]] += 1
    print(L[i3],": ",reg.coef_[i3])
    dico[L[i3]] += 1

print("\n \n \n",dico)

#Supprime les variables qui apparaissent moins de 4 fois du dico
important = {}
for key in dico:
    if dico[key] >= 4:
        important[key] = dico[key]


print("\n \n \n",important)
```
