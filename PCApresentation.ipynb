```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('bodym.csv')
for col in ['subject_id', 'photo_names']:
    df.drop(col, axis=1, inplace=True)
df.head()
```

```python
df_homme = df[df['gender'] == 'male']
df_homme.drop('gender', axis=1, inplace=True)
df_femme = df[df['gender'] == 'female']
df_femme.drop('gender', axis=1, inplace=True)

df_femme.head()
```

```python
from sklearn.decomposition import PCA

# Initialiser le modèle PCA
pca = PCA(n_components=2)  # Nous réduisons à 2 dimensions pour la visualisation

# Ajuster le modèle aux données
principal_components = pca.fit_transform(df_homme)

```

```python
# Afficher les deux composantes principales

plt.scatter(principal_components[:, 0], principal_components[:, 1])
plt.xlabel('Composante principale 1')
plt.ylabel('Composante principale 2')
plt.title('PCA des hommes')
plt.show()
```

```python
# Obtenir la proportion de la variance expliquée
explained_variance_ratio = pca.explained_variance_ratio_
print(f"Proportion de la variance expliquée par la première composante principale: {explained_variance_ratio[0]:.2f}")
print(f"Proportion de la variance expliquée par la deuxième composante principale: {explained_variance_ratio[1]:.2f}")
print(f"Proportion de la variance expliquée par les deux premières composantes principales: {sum(explained_variance_ratio):.2f}")
```

```python
# Initialiser le modèle PCA
pca = PCA(n_components=2)  # Nous réduisons à 2 dimensions pour la visualisation

# Ajuster le modèle aux données
principal_components = pca.fit_transform(df_femme)

# Afficher les deux composantes principales

plt.scatter(principal_components[:, 0], principal_components[:, 1])
plt.xlabel('Composante principale 1')
plt.ylabel('Composante principale 2')
plt.title('PCA des femmes')
plt.show()

# Obtenir la proportion de la variance expliquée
explained_variance_ratio = pca.explained_variance_ratio_
print(f"Proportion de la variance expliquée par la première composante principale: {explained_variance_ratio[0]:.2f}")
print(f"Proportion de la variance expliquée par la deuxième composante principale: {explained_variance_ratio[1]:.2f}")
print(f"Proportion de la variance expliquée par les deux premières composantes principales: {sum(explained_variance_ratio):.2f}")


```
