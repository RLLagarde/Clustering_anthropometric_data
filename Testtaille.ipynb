```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('bodym.csv')
for col in ['subject_id', 'photo_names']:
    df.drop(col, axis=1, inplace=True)
df.head()

```

```python
# Faire une boite à moustache pour chaque colonne de df


for col in df.columns:
    df.boxplot(col)
    plt.show()

#q: A quoi correspondent les points entourés en noir sur les boxplot ?
#r: Ce sont des outliers

#q: C'est quoi un outlier ?
#r: C'est une valeur qui s'écarte de la distribution des
#   autres valeurs
```
